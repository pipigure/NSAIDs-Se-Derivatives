Data Visualization: RF and ANN Model from Dataset Split by RM and KS

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

da = pd.read_csv('/Users/pipi/Documents/研究工作资料/数据/54-pred_1.csv')

train = da.head(28)
test = da.iloc[28:36]
ext = da.tail(16)

y_kstrain = train.iloc[:,0]
y_kstest = test.iloc[:,0]
y_ksext = ext.iloc[:,0]

yp_rftrain = train.iloc[:,1]
yp_rftest = test.iloc[:,1]
yp_rfext = ext.iloc[:,1]

yp_anntrain = train.iloc[:,2]
yp_anntest = test.iloc[:,2]
yp_annext = ext.iloc[:,2]

y1_rstrain = train.iloc[:,3]
y1_rstest = test.iloc[:,3]
y1_rsext = ext.iloc[:,3]

yp1_rftrain = train.iloc[:,4]
yp1_rftest = test.iloc[:,4]
yp1_rfext = ext.iloc[:,4]

yp1_anntrain = train.iloc[:,5]
yp1_anntest = test.iloc[:,5]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.weight'] = 'bold'

plt.figure(figsize=(5, 5))
plt.xlim(1.2, 2.0)
plt.ylim(1.2, 2.0)
x_ticks = np.arange(1.2, 2.1, 0.1)
y_ticks = np.arange(1.2, 2.1, 0.1)
plt.xticks(x_ticks)
plt.yticks(y_ticks)

# 1 row, 2 column, plot 1
plt.subplot(1, 1, 1)
plt.scatter(x=y_kstrain, y=yp_rftrain, alpha=0.4)
plt.scatter(x=y_kstest, y=yp_rftest, c="black", marker='^', alpha=0.4)
plt.legend(("training set", "test set"), loc="lower right")
plt.plot([1.2, 2.0], [1.2, 2.0], "--", color="Black")
plt.title('RF$_{KS}$', fontsize=15)
plt.ylabel('Predicted activity', fontsize= 18,  weight='bold')
plt.xlabel('Experimental activity', fontsize=18,  weight='bold')
plt.savefig("KS-model.svg", format='svg')

```
